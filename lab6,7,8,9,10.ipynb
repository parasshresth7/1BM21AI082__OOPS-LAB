```python

class WrongAge(Exception):
    pass

class AgeInvalid(Exception):
    pass


class Father:
    def __init__(self, age):
        if age < 0:
            raise WrongAge("Age cannot be negative")
        self.age = age


class Son(Father):
    def __init__(self, father_age, son_age):
        super().__init__(father_age)
        if son_age >= father_age:
            raise AgeInvalid("Son's age should be less than father's age")
        self.son_age = son_age

if __name__ == "__main__":
    try:
        father = Father(-3)
        son = Son(50, 25)
        print("Father's age:", father.age)
        print("Son's age:", son.son_age)
    except WrongAge as e:
        print("Error creating Father:", e)
    except AgeInvalid as e:
        print("Error creating Son:", e)
```

```python

class FormulaError(Exception):
    pass


def calculator():
    while True:
        try:
            user_input = input("Enter a formula (e.g., 1 + 1) or 'quit' to exit: ")
            if user_input.lower() == 'quit':
                break
            
            elements = user_input.split()
            if len(elements) != 3:
                raise FormulaError("Invalid input format")
            
            num1 = float(elements[0])
            operator = elements[1]
            num2 = float(elements[2])
            
            if operator not in ['+', '-']:
                raise FormulaError("Invalid operator")
            
            if operator == '+':
                result = num1 + num2
            else:
                result = num1 - num2
            
            print("Result:", result)
        
        except ValueError:
            raise FormulaError("Invalid input. Please enter numeric values")
        
        except FormulaError as e:
            print("Formula Error:", e)

if __name__ == "__main__":
    calculator()
```

```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Set up the screen
screen_width, screen_height = 1280, 960
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("Shapes Drawing")

# Define colors
BLUE = (0, 0, 255)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)

# Draw shapes
def draw_shapes():
    # Clear the screen
    screen.fill((255, 255, 255))
    
    # Draw a blue polygon
    pygame.draw.polygon(screen, BLUE, [(146, 0), (291, 106), (236, 277), (56, 277), (0, 106)])

    # Draw a green line
    pygame.draw.line(screen, GREEN, (60, 300), (120, 300), 10)

    # Draw a green-filled circle
    pygame.draw.circle(screen, GREEN, (300, 50), 20)

    # Draw a black ellipse
    pygame.draw.ellipse(screen, BLACK, pygame.Rect(300, 250, 40, 80), 1)

    # Draw a black rectangle
    pygame.draw.rect(screen, BLACK, pygame.Rect(150, 300, 100, 50))

    # Update the display
    pygame.display.flip()

# Main loop
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()

    # Draw shapes
    draw_shapes()
```

```python
import threading

# Function to check if a number is odd
def is_odd(num):
    return num % 2 != 0

# Function to print the cube of a number
def print_cube(num):
    if is_odd(num):
        print("Cube of", num, "is:", num ** 3)

# Function to print the square of a number
def print_square(num):
    if not is_odd(num):
        print("Square of", num, "is:", num ** 2)

# Main function
def main():
    num = int(input("Enter a number: "))
    
    # Creating threads for printing square and cube
    square_thread = threading.Thread(target=print_square, args=(num,))
    cube_thread = threading.Thread(target=print_cube, args=(num,))
    
    # Starting threads
    square_thread.start()
    cube_thread.start()
    
    # Waiting for both threads to finish
    square_thread.join()
    cube_thread.join()

if __name__ == "__main__":
    main()
```

```python
import threading
import time
import random
from queue import Queue


MARKET_CAPACITY = 10


vegetables = {'tomato': 0, 'carrot': 0, 'lettuce': 0}


customer_queue = Queue()


def produce_vegetables():
    while True:
        for veggie in vegetables:
            vegetables[veggie] += random.randint(1, 5)  
            print(f"Farmer produced {vegetables[veggie]} {veggie}(s)")
        time.sleep(5) 

def buy_vegetables(customer_name, vegetable):
    while True:
        if vegetables[vegetable] > 0:
            print(f"{customer_name} bought {vegetable}")
            vegetables[vegetable] -= 1
            break
        else:
            print(f"{customer_name} is waiting for {vegetable}")
            customer_queue.put((customer_name, vegetable))
            time.sleep(3)  # Wait for 3 seconds before checking again

# Function for the market to serve customers waiting in the queue
def serve_customers():
    while True:
        if not customer_queue.empty():
            customer_name, vegetable = customer_queue.get()
            buy_vegetables(customer_name, vegetable)
            customer_queue.task_done()

# Main function
def main():
    # Create farmer thread to produce vegetables
    farmer_thread = threading.Thread(target=produce_vegetables)
    farmer_thread.start()

    # Create market thread to serve customers
    market_thread = threading.Thread(target=serve_customers)
    market_thread.start()

    # Simulate customers arriving at the market
    for i in range(10):
        customer_name = f"Customer-{i+1}"
        vegetable = random.choice(list(vegetables.keys()))
        customer_thread = threading.Thread(target=buy_vegetables, args=(customer_name, vegetable))
        customer_thread.start()
        time.sleep(random.randint(1, 3))  # Customers arrive randomly every 1-3 seconds

    # Join threads
    farmer_thread.join()
    market_thread.join()

if __name__ == "__main__":
    main()
```

```python
import threading
import time

# Define a lock for synchronization
lock = threading.Lock()

# Function to print numbers from 1 to 5
def function_A():
    for i in range(1, 6):
        with lock:
            print(i)
        time.sleep(0.5)  # Simulate time-consuming task

# Function to print letters from 'A' to 'E'
def function_B():
    for char in 'ABCDE':
        with lock:
            print(char)
        time.sleep(0.5)  # Simulate time-consuming task

# Main function
def main():
    # Create two threads, each for a different function
    thread_A = threading.Thread(target=function_A)
    thread_B = threading.Thread(target=function_B)

    # Start both threads
    thread_A.start()
    thread_B.start()

    # Wait for both threads to complete
    thread_A.join()
    thread_B.join()

if __name__ == "__main__":
    main()
```

```python
class FormulaError(Exception):
    pass


def calculate(formula):
    try:
        elements = formula.split()
        if len(elements) != 3:
            raise FormulaError("Invalid input: Must consist of 3 elements")

        num1 = float(elements[0])
        operator = elements[1]
        num2 = float(elements[2])

        if operator not in ["+", "-"]:
            raise FormulaError("Invalid operator: Use only '+' or '-'")

        if operator == "+":
            result = num1 + num2
        else:
            result = num1 - num2

        print("Result:", result)

    except ValueError:
        raise FormulaError("Invalid input: Unable to convert to float")


# Interactive calculator
while True:
    user_input = input("Enter a formula (or 'quit' to exit): ").strip()

    if user_input.lower() == "quit":
        break

    try:
        calculate(user_input)
    except FormulaError as fe:
        print(f"Error: {fe}")
```

```python
import threading
import time

semaphore = threading.Semaphore(1)

def function_A():
    for num in range(1, 6):
        with semaphore:
            print(num, end=" ")
            time.sleep(0.1)


def function_B():
    for letter in ["A", "B", "C", "D", "E"]:
        with semaphore:
            print(letter, end=" ")
            time.sleep(0.1)

thread_A = threading.Thread(target=function_A)
thread_B = threading.Thread(target=function_B)

thread_A.start()
thread_B.start()

thread_A.join()
thread_B.join()

print("\nProgram terminated.")
```
